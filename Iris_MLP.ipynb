## Dataset used: Iris data

```python
#Importing Necessary libraries
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

```python
from sklearn import preprocessing as prep
```

```python
import keras
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load Data
iris=pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
#Encoding target lables
y = iris.Species
labelEncoder = prep.LabelEncoder()

y = labelEncoder.fit_transform(y).astype(np.float64)

print(y)
```

```python
#Import train test split for dividing data into train and test dataset (Ratio used - 80:20) 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(iris.drop("Species", axis=1), y, test_size = 0.2,\
                                                    random_state=112)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
print(Y_train)
```

```python
#Converting data type to array
x=np.array(X_train)
y=keras.utils.to_categorical(np.array(Y_train),3)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
pd.DataFrame(y).head()
```

There are two ways to build Keras models: sequential and functional.

The sequential API allows you to create models layer-by-layer for most problems. It is limited in that it does not allow you to create models that share layers or have multiple inputs or outputs.

Alternatively, the functional API allows you to create models that have a lot more flexibility as you can easily define models where layers connect to more than just the previous and next layers. In fact, you can connect layers to (literally) any other layer. As a result, creating complex networks such as siamese networks and residual networks become possible.



```python
model=Sequential()
```

**Available activations

1.Relu function

2.Sigmoid Function

3.Softmax Functions

4.Tanh function

5.Elu function 

6.And many more



Here we will use Relu for input side and softmax for output side as this will give us maximum accuracy of predictions

```python
## Create network architecture
model.add(Dense(units=24,input_dim=5))
model.add(Activation('relu'))
model.add(Dense(units=3))
model.add(Activation('softmax'))
```

We will try to calculate cross entropy and accuracy of model to see how well our model can predict

```python
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=100,batch_size=50,validation_split=0.20)
```

```python
## Check for overfit
# Plot training and validation accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

Try to regularate the model Lets see if performance of model is improved or not

```python
from keras.regularizers import l1,l2
```

```python
## Introduce Regularisation
model=Sequential()
model.add(Dense(units=24,input_dim=5,kernel_regularizer=l1(0.001)))
model.add(Activation('relu'))
model.add(Dense(units=3))
model.add(Activation('softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=100,batch_size=50,validation_split=0.20)
```

```python
## Check for overfit
# Plot training and validation accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
## Another way to regularise is to use dropout layer
from keras.layers import Dense, Activation, Dropout
model=Sequential()
## Create network architecture
model.add(Dense(units=40,input_dim=5))
model.add(Activation('sigmoid'))
model.add(Dropout(0.2,seed=100))
model.add(Dense(units=3))
model.add(Activation('softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
history=model.fit(x,y,epochs=100,batch_size=50,validation_split=0.2)
```

```python
## Check for overfit
# Plot training and validation accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
